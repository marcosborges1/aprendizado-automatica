```python
import pandas as pd

# Leitura do arquivo de um arquivo csv qualquer
data_frame = pd.read_csv('data.csv')

# Importar a biblioteca de plotagem
import matplotlib.pyplot as plt

# Definindo o tamanho da janela de plotagem
plt.rcParams['figure.figsize'] = (12.0, 9.0)

# Pegando todos os dados da coluna 0, que é o x. Ficando assim, pegar todas as linhas(:) da coluna 0
x = data_frame.iloc[:,0]
# Pegando todos os dados da coluna 1, que é o x. Ficando assim, pegar todas as linhas(:) da coluna 1
y = data_frame.iloc[:,1]

#Fazendo a plotagem scatter com base nos valores da matriz X, e seu respectivos valores na matriz Y.
plt.scatter(x, y)

# Mostra o scatter plot
plt.show()
```

```python
#Classe Utilidade, para ficar organizado o código
class Utils:
        
        #Método que calcula a gradiente
        def gradient_descendent(y, x, epochs = 1000, alpha=0.0001, w0=0,w1=0): 
            
            #Pega o tamanho da matriz x.
            n = len(x) # or len(y)
            
            #Loop para cada iteração
            for i in range(epochs):
                
                #Função de y previsor (^y)
                yi_predict = w0 + w1*x
                
                #Cálculo dos erros, que são os valores de y real menos os valores de y previsor (ˆy)
                errors = y-y_predict
                
                #Cáculo da dereivada de w0, segundo os slides
                derivate_w0 = (1/n)*sum(errors)
                
                #Cáculo da dereivada de w0, segundo os slides
                derivate_w1 = (1/n)*sum(errors*x)
                
                #Atualização de w0 e w1, segundo os slides
                w0 = w0 + alpha * derivate_w0
                w1 = w1 + alpha * derivate_w1
            #retorna o valor da matriz com os novos valores de Y previsores (ˆy)
            return y_predict
        
        #Método para plotagem, com base no X, Y e Y previsor(^y)
        def make_plot_with_gradient(x, y, y_predict):
            
            # Fazendo a plotagem scatter com base nos valores da matriz X, e seu respectivos valores na matriz Y.
            plt.scatter(x, y)
            plt.title('Gradient Descent')
            
            #Criando a reta de ponstos com base nos valores da matriz X e seus respectivos valores na matriz Y previsor(^y)
            plt.plot(x, y_predict, color="red")
            plt.xlabel('X')
            plt.ylabel('Y')
            plt.show()
            return 
```

```python
# Chama o método de Utils que calcula a gradiente descendente, e obtem os valores da matriz Y previsor (^y)
y_predict = Utils.gradient_descendent(y=y,x=x)

# Chama o método de Utils que plota no gráfico os resultador das matrizes X, Y e Y previsor(^y)
Utils.make_plot_with_gradient(x,y,y_predict)
```
